```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4 , n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# applying ols
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
```

```python
# applying gradienn descent
```

```python
# Starting with assuming we know m = 78.35
# and let b startwith b = 0
```

```python
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red",label="optimal")
plt.plot(X,y_pred,color="green",label="b=0")
plt.legend()
plt.show()
```

```python
# iteration 1
```

```python
m = 78.35
b = 0

slope= -2 * np.sum(y - m*X.ravel() -b )
```

```python
slope
```

```python
lrate = 0.1

step = lrate * slope
step
```

```python
b = b - step
```

```python
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red",label="optimal")
plt.plot(X,y_pred,color="green",label="b=0")
plt.plot(X,y_pred1,color="blue",label="b={}".format(b))
plt.legend()
plt.show()
```

```python
# iteration 2
```

```python
slope2 = -2 * np.sum(y - m*X.ravel() - b )
```

```python
slope2
```

```python
step2 = lrate * slope2
step2
```

```python
b = b - step2
```

```python
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red",label="optimal")
plt.plot(X,y_pred,color="green",label="b=0")
plt.plot(X,y_pred1,color="blue",label="b=20.92")
plt.plot(X,y_pred2,color="purple",label="b={}".format(b))
plt.legend()
plt.show()
```

```python
# iteration 3
```

```python
slope3 = -2 * np.sum(y - m*X.ravel() - b )
```

```python
slope3
```

```python
step3 = lrate * slope3
step3
```

```python
b = b - step3
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red",label="optimal")
plt.plot(X,y_pred,color="green",label="b=0")
plt.plot(X,y_pred1,color="blue",label="b=20.92")
plt.plot(X,y_pred2,color="purple",label="b=25.11")
plt.plot(X,y_pred3,color="black",label="b={}".format(b))
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
lr=0.1

epochs=10

for i in range(epochs):
    lossSlope = -2 * np.sum ( y - m*X.ravel() - b)
    b = b - (lr*lossSlope)
    y_pred = m*X + b
    plt.plot(X,y_pred)

plt.scatter(X,y)    
```
