```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
```

```python
X,y = make_regression(n_samples=100,
    n_features=1,
    n_informative=1,
    n_targets=1,noise=20,random_state=42)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
np.mean(cross_val_score(lr,X,y,scoring="r2",cv=10))
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
```

```python
class GDRegressor:

    def __init__(self,learning_rate,epochs):
        self.m = 100
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,X,y):
        for i in range(self.epochs):
            lossSlopeB = -2 * np.sum ( y - self.m*X.ravel() - self.b)
            lossSlopeM = -2 * np.sum (( y - self.m*X.ravel() - self.b ) * X.ravel())
            self.b = self.b - (self.lr*lossSlopeB)
            self.m = self.m - (self.lr*lossSlopeM)
            # y_pred = self.m*X + self.b
            # plt.plot(X,y_pred)
        print(self.b,self.m)

    def predict(self,X):
        return self.m * X + self.b

gd = GDRegressor(0.001,100)
```

```python
gd.fit(X,y)
```

```python
gd.predict(X)
```
